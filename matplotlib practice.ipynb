```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,2000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('this is my x data')
plt.ylabel('this is my y data')
plt.title('this is my title')
plt.plot(x,y,'--r')
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
colo = np.random.rand(50)
sizes = np.random.rand(50)*500
plt.scatter(x,y,c=colo,s = sizes,alpha=.90)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('random data')
```
